# Exploratory Data Analysis (Case Study: Titantic Dataset)

In this notebook, we will perform exploratory data analysis (EDA) on the Titanic dataset. Our goals are to:
- Calculate key descriptive statistics such as mean, median, mode, range, variance, and standard deviation.
- Identify potential biases and assess the completeness of the data.

Let's dive in and start exploring the dataset!


## Importing Libraries and Loading the Dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic.head()
```

## Understanding the Titanic Dataset

The Titanic dataset contains information about the passengers on the Titanic. Let's start by taking a look at the first few rows of the dataset to understand its structure and the types of data it contains.

The dataset includes features such as:
- `survived`: Survival status (0 = No, 1 = Yes)
- `pclass`: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd)
- `sex`: Gender
- `age`: Age in years
- `sibsp`: Number of siblings/spouses aboard
- `parch`: Number of parents/children aboard
- `fare`: Ticket fare
- `embarked`: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

We'll use these features to calculate descriptive statistics and assess data quality.


## Mean, Median and Mode

```python
# Calculating Mean, Median, and Mode
mean_age = titanic['age'].mean()
median_age = titanic['age'].median()
mode_age = titanic['age'].mode()[0]

mean_fare = titanic['fare'].mean()
median_fare = titanic['fare'].median()
mode_fare = titanic['fare'].mode()[0]

# Display the results
print(f"Mean Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")

print(f"Mean Fare: {mean_fare}")
print(f"Median Fare: {median_fare}")
print(f"Mode Fare: {mode_fare}")
```

The mean age of the passengers provides a general sense of the average age, but the median might be more robust if the data includes outliers. The mode can indicate if there's a common age among passengers.

Next, let's explore the range, variance, and standard deviation for these features.


```python
# Calculate range, variance, and standard deviation for 'age' and 'fare'
range_age = titanic['age'].max() - titanic['age'].min()
variance_age = titanic['age'].var()
std_dev_age = titanic['age'].std()

range_fare = titanic['fare'].max() - titanic['fare'].min()
variance_fare = titanic['fare'].var()
std_dev_fare = titanic['fare'].std()

# Display the results
print(f"Range of Age: {range_age}")
print(f"Variance of Age: {variance_age}")
print(f"Standard Deviation of Age: {std_dev_age}")

print(f"Range of Fare: {range_fare}")
print(f"Variance of Fare: {variance_fare}")
print(f"Standard Deviation of Fare: {std_dev_fare}")
```

## Visualizing Descriptive Statistics

Visualizing the distribution of data helps us understand its characteristics. Let's create some histograms and box plots to explore the distribution of ages and fares.


```python
# Visualizing Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'].dropna(), kde=True)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Visualizing Fare Distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic['fare'].dropna(), kde=True)
plt.title('Fare Distribution of Titanic Passengers')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

## Assessing Data Completeness

Missing data can lead to biased results if not handled properly. Let's identify any missing values in our dataset and visualize them.


```python
# Checking for Missing Data
missing_values = titanic.isnull().sum()
print("Missing Values in the Dataset:\n", missing_values)

# Visualizing Missing Data
sns.heatmap(titanic.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values in the Titanic Dataset')
plt.show()
```

## Identifying Potential Biases

Bias in the dataset can affect the outcome of our analysis. Let's explore some visualizations to identify potential biases, such as class distribution and survival rates by gender.


```python
# Distribution of passengers by class
sns.countplot(x='pclass', data=titanic)
plt.title('Passenger Class Distribution')
plt.show()

# Survival rate by gender
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.show()
```
